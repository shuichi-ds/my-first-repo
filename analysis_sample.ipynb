```python
import pandas as pd
import matplotlib.pyplot as plt

# サンプルデータ
data = {'Month': ['Jan', 'Feb', 'Mar'], 'Sales': [100, 120, 90]}
df = pd.DataFrame(data)

# データ表示
print(df)

# グラフ描画
plt.plot(df['Month'], df['Sales'])
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
